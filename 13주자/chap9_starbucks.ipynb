```python
import pandas as pd
SB = pd.read_csv('C:/Users/MinJu/Desktop/과제/빅데이터/13week/starbucks.csv', encoding = 'CP949',engine= 'python')
SB.head()
```

```python
addr = []
for address in SB.address :
    addr.append(str(address).split())
addr
```

```python
import folium
```

```python
map_osm = folium.Map(location=[37.559978,126.975291], zoom_start=16)
```

```python
for i, store in SB.iterrows():
    folium.Marker(location=[store['위도'],store['경도']], popup= store['store'], icon=folium.Icon(icon_color='red', icon='heart')).add_to(map_osm)
```

```python
map_osm.save('C:/Users/MinJu/Desktop/과제/빅데이터/13week/starbucks.html')
```
